### Finding NaN in data

We find out that there is value of NaN in our data. We can check how many NaN in our data first to determind how these NaN will influence our model. In the table below, we find out that there are most NaN in climate feature, which is 9.69%, and less in GDP feature. Finally, there are 110 NaN in our data totally, it may cause the inaccuracy of our training model and we will eliminated these NaN value in the next section.

```python
total = df.isnull().sum()[df.isnull().sum() != 0].sort_values(ascending = False)
percent = pd.Series(round(total/len(df)*100,2))
pd.concat([total, percent], axis=1, keys=['total_missing', 'percent'])
```

```python
df.isnull().sum().sum()
```

### Eliminate the NaN

An easy way to eliminate the NaN value, we can just modify them by the feature's mean value. The reason is because we don't actually know these value, so just let them to be median value of the column data for less affect. After executing our step, we can just easily check there are no NaN in our data now.

```python
for col in df.columns.values:
    if df[col].isnull().sum() == 0:
        continue
    if(col == 'Country' or col == 'Region'):
        continue
    df[col].loc[df[col].isnull()] = df[col].median()
    
```

```python
df.isnull().sum()
```

### Features

The figures below show that the scatterplot of GDP versus other features. Because we will use linear regression to train our model, so we can just briefly pick the feature candidates which have linear relationship with our target, GDP. As the result, obviously 'Phones' has the highest corelation value, then is 'Service'. The number of the features is the matter of opinion, which is the topic of overfitting and underfitting.

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()
for col in df.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = df['GDP ($ per capita)'].corr(df[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP ($ per capita)', data=df,
                   ax=axes[i,j], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```

```python
gdp_corr_seq = (df.corr()['GDP ($ per capita)']).sort_values(ascending=False).drop('GDP ($ per capita)')
gdp_corr_seq
```
